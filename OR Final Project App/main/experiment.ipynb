```python
from algorythm import Population, Fixtures, Genetic, Team, Match
from copy import deepcopy
import matplotlib.pyplot as plt
import psutil
import time
```

```python
Genn1 = Genetic(40, 30)

Genn1.create_population()

best_result1 = [None for _ in range(Genn1.generation_size)]
best_result2 = [None for _ in range(Genn1.generation_size)]
best_result3 = [None for _ in range(Genn1.generation_size)]
best_result4 = [None for _ in range(Genn1.generation_size)]

best_result1[0] = max(Genn1.population)
best_result2[0] = max(Genn1.population)
best_result3[0] = max(Genn1.population)
best_result4[0] = max(Genn1.population)

test1 = deepcopy(Genn1)
test2 = deepcopy(Genn1)
test3 = deepcopy(Genn1)
test4 = deepcopy(Genn1)

test1.tournament_selection(10, 0.05)
test2.tournament_selection(10, 0.05)
test3.tournament_selection(10, 0.05)
test3.tournament_selection(10, 0.05)

test1.cross()
test2.cross()
test3.cross()
test4.cross()

for i in range(1, test1.generation_size):
    new_gen1 = deepcopy(test1)
    new_gen1.tournament_selection(5, 0.05)
    new_gen1.cross()
    new_gen1.mutate1(0.1, 0.1)
    new_gen1.mutate2(0.1, 0.1)
    new_gen1.mutate3(0.1, 0.1)
    new_gen1.mutate4(0.1)
    best_result1[i] = max(new_gen1.population)
    test1 = new_gen1
    del new_gen1

for i in range(1, test2.generation_size):
    new_gen2 = deepcopy(test2)
    new_gen2.tournament_selection(10, 0.05)
    new_gen2.cross()
    new_gen2.mutate1(0.1, 0.1)
    new_gen2.mutate2(0.1, 0.1)
    new_gen2.mutate3(0.1, 0.1)
    new_gen2.mutate4(0.1)
    best_result2[i] = max(new_gen2.population)
    test2 = new_gen2
    del new_gen2

for i in range(1, test3.generation_size):
    new_gen3 = deepcopy(test3)
    new_gen3.tournament_selection(20, 0.05)
    new_gen3.cross()
    new_gen3.mutate1(0.1, 0.1)
    new_gen3.mutate2(0.1, 0.1)
    new_gen3.mutate3(0.1, 0.1)
    new_gen3.mutate4(0.1)
    best_result3[i] = max(new_gen3.population)
    test3 = new_gen3
    del new_gen3

for i in range(1, test4.generation_size):
    new_gen4 = deepcopy(test4)
    new_gen4.tournament_selection(10, 0.05)
    new_gen4.cross()
    new_gen4.mutate1(0.1, 0.1)
    new_gen4.mutate2(0.1, 0.1)
    new_gen4.mutate3(0.1, 0.1)
    new_gen4.mutate4(0.1)
    best_result4[i] = max(new_gen4.population)
    test4 = new_gen4
    del new_gen4
```

```python
print(best_result1)
idx1 = list(range(1, len(best_result1) + 1))
y_wartosci1 = [obiekt.objective_function() for obiekt in best_result1]
best1 = max(best_result1).objective_function()
```

```python
plt.plot(idx1, y_wartosci1)
plt.xlabel('Iteracja')
plt.ylabel('Wartość funkcji celu')
plt.title('Wykres funkcji celu')
plt.xlim([1, test1.generation_size])
plt.axhline(y=best1, color='r', linestyle='--', label=f'Maks: {best1}')
plt.text(1, best1, f' Maks: {round(best1, 2)}', color='r', fontsize=10, va='bottom', ha='left')

plt.grid()
plt.show()
```

```python
def your_function(param):
    Genn1 = Genetic(param, 30)
    Genn1.create_population()

    best_result1 = [None for _ in range(Genn1.generation_size)]

    Genn1.tournament_selection(10, 0.05)
    Genn1.cross()

    for i in range(1, Genn1.generation_size):
        new_gen1 = deepcopy(Genn1)
        new_gen1.tournament_selection(5, 0.05)
        new_gen1.cross()
        new_gen1.mutate1(0.1, 0.1)
        best_result1[i] = max(new_gen1.population)
        Genn1 = new_gen1
        del new_gen1

    memory_usage = psutil.Process().memory_info().rss
    return memory_usage
```

```python
def generate_memory_plot():
    params = list(range(20, 91))  # Lista różnych wartości parametru
    memory_usages = []

    for param in params:
        memory_usage = your_function(param)
        memory_usages.append(memory_usage)

    # Tworzenie wykresu
    plt.plot(params, memory_usages, marker='o')
    plt.xlabel('Wielkość pokolenia')
    plt.ylabel('Zużycie pamięci (w bajtach)')
    plt.title('Zużycie pamięci w zależności od parametru')
    plt.grid()
    plt.show()


generate_memory_plot()
```

```python
def your_function2(param):
    start_time = time.time()

    Genn1 = Genetic(param, 30)
    Genn1.create_population()

    best_result1 = [None for _ in range(Genn1.generation_size)]

    Genn1.tournament_selection(10, 0.05)
    Genn1.cross()

    for i in range(1, Genn1.generation_size):
        new_gen1 = deepcopy(Genn1)
        new_gen1.tournament_selection(5, 0.05)
        new_gen1.cross()
        new_gen1.mutate1(0.1, 0.1)
        best_result1[i] = max(new_gen1.population)
        Genn1 = new_gen1
        del new_gen1

    end_time = time.time()

    execution_time = end_time - start_time
    return execution_time

def generate_time_plot():
    params = list(range(20, 91))  # Lista różnych wartości parametru
    execution_times = []

    for param in params:
        execution_time = your_function2(param)
        execution_times.append(execution_time)

    # Tworzenie wykresu
    plt.plot(params, execution_times, marker='o')
    plt.xlabel('Wielkość pokolenia')
    plt.ylabel('Czas wykonania (w sekundach)')
    plt.title('Czas wykonania w zależności od parametru')
    plt.grid()
    plt.show()

generate_time_plot()
```
